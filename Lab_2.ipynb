```python
import pandas as pd

df = pd.read_csv(r"C:\Users\itsme\Downloads\lab_2.csv")
df['voltage'] = (df['adc_raw'] * 5.0)/1023
Vcc = 5.0
R_fixed = 1000.0  # 1 kΩ
V = df["voltage"] # volts at A0
df["R_therm"] = R_fixed * (Vcc / V - 1.0)

R25 = 3300.0
R = df['R_therm']
A1  = 0.00335401643468053
B1  = 0.00025698501802
C1  = 0.0000026201306709
D1  = 0.000000063830907998

L = np.log(R / R25)
invT = A1 + B1*L + C1*(L**2) + D1*(L**3)  
df["T_K"] = 1.0 / invT
df['time_seconds']  = df['time_ms']/1000
df

plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
plt.title('Time vs Temperature')
plt.plot(df['time_seconds'], df['T_K'])






```
