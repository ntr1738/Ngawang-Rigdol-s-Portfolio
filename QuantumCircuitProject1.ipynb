```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3)
qc.h(0)
qc.h(1)
qc.cx(1, 2)
qc.cx(0, 1)
qc.x(0)
qc.x(1)
qc.id(2)
qc.x(2)
qc.measure_all()
qc.draw(output='mpl')



```

```python
from qiskit_aer import AerSimulator  

simulator = AerSimulator()
x = simulator.run(qc, shots=1000000)
results = x.result()
#print(results)
counts = results.get_counts()
print(counts)









```
