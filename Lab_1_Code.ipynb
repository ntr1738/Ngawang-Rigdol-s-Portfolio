```python
# Measurements and Plot

import matplotlib.pyplot as plt

analog_reading = [387,388,388,388,388,387,388,389,388,387,388,387,387,388,388]
v_out = []
for i in analog_reading:
    v_out_vals = round((i/1023)*5,3)
    v_out.append(v_out_vals)

r_known = 330
v_in = 5
r_unknown = []
for j in v_out:
    r_known_vals = round((j*r_known)/(v_in-j),3)
    r_unknown.append(r_known_vals)

print(r_unknown)

plt.title('Unknown Resistance vs Output Voltage')
plt.plot(r_unknown,v_out)
plt.xlabel('Resistance of Unknown (Ohms)')
plt.ylabel('Output Voltage (V)')

# Save as PDF for LaTeX
plt.savefig("voltage_divider.pdf")
plt.show()


```

```python
# Precision and Accuracy
v_out_average = round(sum(v_out)/float(len(v_out)),3)
#print(v_out_average)
ADC_step = round(v_in/1023,3)
#print(ADC_step)
r_unknown_high = r_known*(v_out_average+ADC_step)/(v_in-(v_out_average+ADC_step))
#print(round(r_unknown_high,3))
r_unknown_low = r_known*(v_out_average-ADC_step)/(v_in-(v_out_average-ADC_step))
#print(round(r_unknown_low,3))
r_error_high = round((r_unknown_high - r_unknown_low)/2,3)
r_error_low = round((r_unknown_high - r_unknown_low)/2,3) * -1
#print(r_error_high)
#print(r_error_low)
r_unknown_average = round(float(sum(r_unknown)/len(r_unknown)),3)
#print(r_unknown_average)
precision_1 = r_unknown_average + r_error_high
precision_2 = r_unknown_average + r_error_low
print(precision_1) 
print(precision_2)
print(r_unknown_average)
print(r_error_high)



```
